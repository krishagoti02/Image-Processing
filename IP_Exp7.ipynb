# Erosion : <br>
In the erosion operation, if for EVERY pixel that is true (1 / shaded) in the structuring element, the corresponding pixel in the image is also true (1/shaded), then the output pixel is set to true (1/shaded). Otherwise, the output pixel is set to false (0/unshaded).

#  Python code for performing erosion

```python
#Read the image for erosion squares.tif
img = cv2.imread('erosion.png',0)
plt.imshow(img,cmap='gray')


r,c = img.shape
erode=np.zeros((r,c),dtype=int)
for i in range(1,r-1):
  for j in range(1,c-1):
    temp=[img[i-1,j-1],img[i-1,j],img[i-1,j+1],img[i,j-1],img[i,j],img[i,j+1],img[i+1,j-1],img[i+1,j],img[i+1,j+1]]
    erode[i,j]=min(temp)

plt.subplot(121)
plt.imshow(img,cmap='gray')
plt.title(' Image')
plt.subplot(122)
plt.imshow(erode,cmap='gray')
plt.title('Eroded Image')
plt.tight_layout()
```

```python
img = cv2.imread('squares.tif',0)
plt.imshow(img,cmap='gray')

r,c = img.shape
erode=np.zeros((r,c),dtype=int)
for i in range(1,r-1):
  for j in range(1,c-1):
    temp=[img[i-1,j-1],img[i-1,j],img[i-1,j+1],img[i,j-1],img[i,j],img[i,j+1],img[i+1,j-1],img[i+1,j],img[i+1,j+1]]
    erode[i,j]=min(temp)

plt.subplot(121)
plt.imshow(img,cmap='gray')
plt.title(' Image')
plt.subplot(122)
plt.imshow(erode,cmap='gray')
plt.title('Eroded Image')
plt.tight_layout()
```

```python
img = cv2.imread('Image_for_dilation.png',0)
plt.imshow(img,cmap='gray')

r,c = img.shape
erode=np.zeros((r,c),dtype=int)
for i in range(1,r-1):
  for j in range(1,c-1):
    temp=[img[i,j-1],img[i,j],img[i,j+1]]
    erode[i,j]=min(temp)

plt.subplot(121)
plt.imshow(img,cmap='gray')
plt.title(' Image')
plt.subplot(122)
plt.imshow(erode,cmap='gray')
plt.title('Eroded Image')
plt.tight_layout()
```

Dilation for Image

```python
img = cv2.imread('Image_for_dilation.png',0)
plt.imshow(img,cmap='gray')


r,c = img.shape
dilate=np.zeros((r,c),dtype=int)
for i in range(1,r-1):
  for j in range(1,c-1):
    temp=[img[i-1,j-1],img[i-1,j],img[i-1,j+1],img[i,j-1],img[i,j],img[i,j+1],img[i+1,j-1],img[i+1,j],img[i+1,j+1]]
    dilate[i,j]=max(temp)

plt.subplot(121)
plt.imshow(img,cmap='gray')
plt.title(' Image')
plt.subplot(122)
plt.imshow(dilate,cmap='gray')
plt.title('Dilation Image')
plt.tight_layout()
```

Erosion using built in function

```python
#Erosion using CV2 built in function to obtain structuring element and perform erosion
SE = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
erodeout = cv2.erode(img,SE)
dilateout = cv2.dilate(img,SE)

#Plot the original image, image obtained by hard coding and image obtained by built in function
plt.figure(figsize = (20,20))

plt.subplot(131)
plt.imshow(img,cmap='gray')
plt.title(' Image')
plt.subplot(132)
plt.imshow(erodeout,cmap='gray')
plt.title('Erode output')
plt.subplot(133)
plt.imshow(dilateout,cmap='gray')
plt.title('Dilated output')
```

# Dilation

In the dilation operation, if for ANY pixel that is true (1/shaded) in the structuring element, the corresponding pixel in the image underneath is also true (1/shaded) , then the output pixel is set to true. Otherwise, the output pixel is set to false.

image.png

Morphological opening of an image is basically erosion followed by dilation:
AB=OPEN(A,B)=D(E(A))


In the example below, application of opening and closing is shown to filter the noise in the given fingerprint image. First the image is opened using erosion followed by dilation operation using the given structuring element. Then on the opened image, closing operation of dilation followed by erosion operation is performed.

It is seen that after applying the opening followed by closing operation, the noise on the finger print image is filtered.

Morphological Opening Operation

```python
#Function for erosion
img=cv2.imread('noisy_fingerprint.png')
SE = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
o1 = cv2.erode(img,SE)
o2 = cv2.dilate(o1,SE)

plt.figure(figsize = (20,20))

plt.subplot(131)
plt.imshow(img,cmap='gray')
plt.title(' Image')
plt.subplot(132)
plt.imshow(o1,cmap='gray')
plt.title('E(A,B)')
plt.subplot(133)
plt.imshow(o2,cmap='gray')
plt.title('O(A,B)')
```

Morphological Closing Operation

```python
img=cv2.imread('noisy_fingerprint.png')
SE = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
o1 = cv2.dilate(img,SE)
o2 = cv2.erode(o1,SE)

plt.figure(figsize = (20,20))

plt.subplot(131)
plt.imshow(img,cmap='gray')
plt.title(' Image')
plt.subplot(132)
plt.imshow(o1,cmap='gray')
plt.title('E(A,B)')
plt.subplot(133)
plt.imshow(o2,cmap='gray')
plt.title('O(A,B)')

```

Import Libraries

```python
#import cv2, numpy, matplotlib
import cv2
from skimage import io
from skimage.color import rgb2gray
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

Theory:<br>
The Sobel operator performs a 2-D spatial gradient measurement on an image and so emphasizes regions of high spatial frequency that correspond to edges.


```python
# load the image house.tiff and read the same
img = cv2.imread('house.tif',0)
plt.imshow(img,cmap='gray')
plt.title('Original image')

```

Sobel operator for horizontal edges <br>
image.png

```python
#Detect horizontal edges using in built convolution function
mask = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
gx=signal.convolve(img,mask)

plt.imshow(gx,cmap='gray',vmin=0,vmax=255)
plt.title('Horizontal Image')

```

Sobel operator for vertical edges <br>
image.png

```python
#Detect vertical edges using inbuilt function for convolution
mask = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
gy=signal.convolve(img,mask)

plt.imshow(gy,cmap='gray',vmin=0,vmax=255)
plt.title('Vertical Image')


```

Detecting vertical and horizontal edges using |gx|+|gy|

```python
#Detecting vertical and horizontal edges

output = gx+gy
plt.imshow(output,cmap='gray',vmin=0,vmax=255)


```

# Students have to perform all the above steps for jetplane.tif image

```python
# load the jetplane.tif image and read the same
img = cv2.imread('jetplane.tif',0)
plt.imshow(img,cmap='gray')
```

```python
#Detect horizontal edges using in built convolution function

mask = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
gx=signal.convolve(img,mask)

plt.imshow(gx,cmap='gray',vmin=0,vmax=255)
plt.title('Horizontal Image')
```

```python
#Detect vertical edges using inbuilt function for convolution


mask = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
gy=signal.convolve(img,mask)

plt.imshow(gy,cmap='gray',vmin=0,vmax=255)
plt.title('Vertical Image')
```

```python
# Detecting Backward Diagonal with inbuilt convolve
mask = np.array([[0,-1,-2],[1,0,-1],[2,1,0]])
gy2=signal.convolve(img,mask)

plt.imshow(gy2,cmap='gray',vmin=0,vmax=255)
plt.title('Backward Image')
```

Forward diagonal<br>
 diagonal edges <br>
image.png


```python
# Detecting Forward Diagonal with inbuilt convolve
mask = np.array([[-2,-1,0],[-1,0,1],[0,1,2]])
gy1=signal.convolve(img,mask)

plt.imshow(gy1,cmap='gray',vmin=0,vmax=255)
plt.title('Forward Image')
```

# Edge in all four directions <br>

```python
#Horizontal, vertical and diagonal edges
output = gx+gy+gy2+gy1
plt.imshow(output,cmap='gray',vmin=0,vmax=255)
plt.title(' Image')
```

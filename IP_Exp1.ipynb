
To convert a color (RGB) image to Grayscale

This example converts an image with RGB channels into an image with a single grayscale channel.

Y = 0.2125 R + 0.7154 G + 0.0721

These weights are used by CRT phosphors as they better represent human perception of red, green and blue than equal weights

```python
from skimage import io
from skimage.color import rgb2gray
img = io.imread('cat.png') #color image
gray_img = rgb2gray(img) # convert to gray scale
plt.subplot(121)
io.imshow(img)
plt.title('Original image')
plt.subplot(122)
io.imshow(gray_img)
plt.title('Grayscale image')
plt.show()
```

While converting a color to gray image you can loose some information Ishihara plates are used to check color blindness

```python
plt.figure(figsize=(10,10)) # increase the size
img = io.imread('ishihara.jpg') #color image
gray_img = rgb2gray(img) # convert to gray scale
plt.subplot(121)
io.imshow(img)
plt.title('Original image')
plt.subplot(122)
io.imshow(gray_img)
plt.title('Grayscale image')
plt.show()
```

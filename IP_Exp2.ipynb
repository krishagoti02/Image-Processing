# Theory: <br>
A resolution can be defined as the total number of pixels in an image. Size of an image is given by N X M x m where m is the no. of bits used to represent each pixel, N represents number of rows and M represents number of columns. <br>

Spatial resolution: Number of rows and columns in an image specifies the spatial resolution of an image. If N=8 and M=8 then there will be 8 X8 =64 pixels in that image.<br>

Grey level resolution: The range of grey level in any image depends on number of bits per pixel. Example, if m=8 then there will be 2^8 = 256 shades of grey ranging from 0 to 255. <br>

#Downsampling : <br>
In the down-sampling technique, the number of pixels in the given image is reduced depending on the sampling frequency. Due to this, the resolution and size of the image decrease.<br>

If sampling frequency =2, output every 2nd element row wise and column wise

Example: <br>
image.png

```python
import warnings
# Import libraries
from skimage import io
import numpy as np
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')
```

```python
# read the cameraman.tif image as img1

img = io.imread('cameraman.tif')

# find the number of rows and columns of this image
[r,c]=img.shape
f=2 #sampling freq

# create a array of zeros of half the image size with variable namse img2
dsimg = np.zeros((r//f,c//f),dtype=np.int)
dsimg.shape
```

```python
# Provide the sampling factor
#write a for loop and fill the alternate pixels of img1 in img2
for i in range(0,r,f):
  for j in range(0,c,f):
    dsimg[i//f,j//f] = img[i,j]
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(dsimg,cmap='gray')
plt.title('Downsampled image')

#Plot the original and the down sampled image
print('Original image size=',img.shape)
print('Downsampled image size=',dsimg.shape)
```

#Upsampling <br>
The number of pixels in the down-sampled image can be increased by using up-sampling interpolation techniques. The up-sampling technique increases the resolution as well as the size of the image.<br>


Perform upsampling by zero stuffing

#Example <br>
image.png

```python
# Up sampling
#make an array of zeros of size of original image. name this array as img3.
img3 = np.zeros((r,c),dtype=np.int)
for i in range (0,r-1,f):
  for j in range(0,c-1,f):
    # fill the alternate vaue of the imge 3 array with the value of img2
    img3[i,j]=dsimg[i//f,j//f]
plt.subplot(1,2,1)
plt.imshow(dsimg,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(img3,cmap='gray')

# plot the upsampled image
plt.title('Upsampled image')
```

# Amplitude (Intensity) Resolution <br>
It refers to the number of intensity levels used to represent the image.
The more intensity levels used, the finer the level of detail discernible in an image. Intensity level resolution is usually given in terms of the number of bits used to store each intensity level.


To reduce the intensity levels from 8 bit to b bit
This can be done as follows:
if x is the original pixel (of 8 bit image) and <br>
   y is the new pixel value (of b bit image) <br>

If 255 --> 2^b <br>
   x --> y <br>

   Thus new pixel value <br>
   y = (x * 2^b) / 255

```python
# read the cameraman.tiff file as imag1
from skimage import io
import matplotlib.pyplot as plt

img=io.imread('cameraman.tif')

# make a array of zeros of size same as the original image img1
I= img.copy()
r,c = img.shape
plt.figure(figsize=(15,15))
# write a for loop and replace each pixel value of img1 with img2*2^b/255
b=int(input('Enter no of bits='))
for i in range(r):
  for j in range(c):
    I[i,j]= I[i,j]* (2**b)/255

# plot the original image and the modified image
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title('8 bit image')
plt.subplot(1,2,2)
plt.imshow(I,cmap='gray')
plt.title('2 bit image')
```
